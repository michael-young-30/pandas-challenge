```python
import pandas as pd

file="./purchase_data.csv"
```

```python
file
purchase_df=pd.read_csv(file)
df=purchase_df
df
```

```python
df.count()
```

```python
total_players=len(df["SN"].unique())
total_players_df=pd.DataFrame({"Total Players":[total_players]})
total_players_df
```

```python
unique_items=len(df["Item Name"].unique())
unique_items

average_price=df["Price"].mean()
rounded_average_price=round(average_price,2)
rounded_average_price

total_purchases=df['Purchase ID'].count()
total_purchases

total_revenue=df["Price"].sum()
total_revenue
```

```python
print(f'Player Count: {total_players}')
purchasing_analysis_df = pd.DataFrame({
    "Number of Unique Items":[unique_items],
    "Average Purchase Price":[rounded_average_price],
    "Total Number of Purchases":[total_purchases],
    "Total Revenue":[total_revenue]    
    
})

purchasing_analysis_df
```

```python
male_df=df.loc[df["Gender"]=="Male",:]
male_df
total_male_players=len(male_df["SN"].unique())
total_male_players

female_df=df.loc[df["Gender"]=="Female",:]
total_female_players=len(female_df["SN"].unique())
total_female_players

other_df=df.loc[df["Gender"]=="Other / Non-Disclosed",:]
other_df
total_other_players=len(other_df["SN"].unique())
total_other_players

percent_male_players=round((total_male_players/total_players)*100,0)
percent_female_players=round((total_female_players/total_players)*100,0)
percent_other_players=round((total_other_players/total_players)*100,0)

print(f'{percent_male_players},{percent_female_players},{percent_other_players}')



print(f'Total Players by Gender\n'
f'Male: {total_male_players}({percent_male_players}%)\n'
f'Female: {total_female_players}({percent_female_players}%)\n'
f'Other/Non-disclosed: {total_other_players}({percent_other_players}%)\n')
```

```python
#Male

male_df=df.loc[df["Gender"]=="Male",:]
male_df

average_price_male= male_df["Price"].mean()
rounded_average_price_male=round(average_price_male,2)
rounded_average_price_male

total_purchases_male= male_df['Purchase ID'].count()
total_purchases_male

total_revenue_male=male_df["Price"].sum()
total_revenue_male

average_per_male=round(total_revenue_male/total_male_players,2)
average_per_male

#Female

female_df=df.loc[df["Gender"]=="Female",:]
female_df

average_price_female= female_df["Price"].mean()
rounded_average_price_female=round(average_price_female,2)
rounded_average_price_female

total_purchases_female= female_df['Purchase ID'].count()
total_purchases_female

total_revenue_female=female_df["Price"].sum()
total_revenue_female

average_per_female=round(total_revenue_female/total_female_players,2)
average_per_female


# Other
other_df=df.loc[df["Gender"]=="Other / Non-Disclosed",:]
other_df

average_price_other= other_df["Price"].mean()
rounded_average_price_other=round(average_price_other,2)
rounded_average_price_other

total_purchases_other= other_df['Purchase ID'].count()
total_purchases_other

total_revenue_other=other_df["Price"].sum()
total_revenue_other

average_per_other=round(total_revenue_other/total_other_players,2)
average_per_other


gender_summary_df = pd.DataFrame({
    "Gender":["Male","Female","Other/Non-disclosed"],
    "Purchase Count":[total_purchases_male,total_purchases_female,total_purchases_other],
    "Average Purchase Price":[rounded_average_price_male,rounded_average_price_female,rounded_average_price_other],
    "Total Revenue":[total_revenue_male,total_revenue_female,total_revenue_other],
    "Averare Purchase Total per Person by Gender":[average_per_male,average_per_female,average_per_other]
    })
gender_summary_df

gender_summary_df["Average Purchase Price"]=gender_summary_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
gender_summary_df["Total Revenue"]=gender_summary_df["Total Revenue"].astype(float).map("${:,.2f}".format)
gender_summary_df["Averare Purchase Total per Person by Gender"]=gender_summary_df["Averare Purchase Total per Person by Gender"].astype(float).map("${:,.2f}".format)

gender_summary_df
```

```python
max_age=df["Age"].max()
print(max_age)
min_age=df["Age"].min()
print(min_age)

bins = [0,9.90,14.90,19.90,24.90,29.90,34.90,39.90,44.90,49.90,54.90]
group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40-44","45-49","50+"]

df["Age Summary"] = pd.cut(df["Age"], bins, labels=group_names, include_lowest=True)

age_group_df=df.groupby(["Age Summary"])

total_age_players=age_group_df["SN"].nunique()
total_age_players

average_price_age= age_group_df["Price"].mean()
rounded_average_price_age=round(average_price_age,2)
rounded_average_price_age

total_purchases_age= age_group_df['Purchase ID'].count()
total_purchases_age

total_revenue_age=age_group_df["Price"].sum()
total_revenue_age

average_per_age=round(total_revenue_age/total_age_players,2)
average_per_age

age_summary =pd.DataFrame({
    "Purchase Count":total_age_players,
    "Average Purchase Price":rounded_average_price_age,
    "Total Revenue":total_revenue_age,
    "Averare Purchase Total per Person by age group":average_per_age })
age_summary

```

```python
sn_group_df=df.groupby(['SN'])
total_sn_purchase=sn_group_df["Purchase ID"].count()
average_sn=sn_group_df["Price"].mean()
total_purchase_value=sn_group_df["Price"].sum()

top_spenders_df =pd.DataFrame({
    "Purchase Count":total_sn_purchase,
    "Average Purchase Price":average_sn,
    "Total Purchase Value":total_purchase_value })

top_5_spenders_df=top_spenders_df.sort_values("Total Purchase Value", axis=0, ascending=False).head(5)

top_5_spenders_df["Average Purchase Price"]=top_5_spenders_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
top_5_spenders_df["Total Purchase Value"]=top_5_spenders_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)

top_5_spenders_df
```
